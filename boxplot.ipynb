```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file_path = r"C:\Users\Admin\Downloads\ADI_Data_with_Additional_Independent_Variables.csv"
adi_data = pd.read_csv(file_path)
print(adi_data.head())
```

```python
file_path = r"C:\Users\Admin\Downloads\AMD.csv"
amd_data = pd.read_csv(file_path)
print(amd_data.head())
```

```python
file_path = r"C:\Users\Admin\Downloads\AVGO.csv"
avgo_data = pd.read_csv(file_path)
print(avgo_data.head())
```

```python
plt.figure(figsize=(6, 8))
plt.boxplot(adi_data['Adj_Close'], vert=True, patch_artist=True)
plt.title('Boxplot of Adj_Close for ADI (Vertical Orientation)')
plt.ylabel('Adj_Close')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(6, 8))
plt.boxplot(amd_data['Adj_Close'], vert=True, patch_artist=True)
plt.title('Boxplot of Adj_Close for AMD (Vertical Orientation)')
plt.ylabel('Adj_Close')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(6, 8))
plt.boxplot(avgo_data['Adj_Close'], vert=True, patch_artist=True)
plt.title('Boxplot of Adj_Close for AVGO (Vertical Orientation)')
plt.ylabel('Adj_Close')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
